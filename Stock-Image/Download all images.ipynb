```python
data = []

for sp in soup.find_all('div', class_ = 'container'):
    print(sp)
    img_link = sp.find('img').get('src')
    tags     = sp.find('span', class_ = 'tag-color').text[7:]
    likes    = sp.find('div', class_ = 'likes-comments').find_all('span')[0].text.strip()[:-6]
    comments = sp.find('div', class_ = 'likes-comments').find_all('span')[1].text.strip()[:-9]

    data.append([img_link, tags, likes, comments])
    #print(img_link)
```

```python
df = pd.DataFrame(data, columns = ['img_link', 'tags', 'likes', 'comments'])
df.head()
```

## generate dynamic names

```python
ref = 10000

for i in range(1,11):
    print('A' + str(i + ref)[1:])
```

## define download function 

```python
def download(link, img_path ):
    
    res = requests.get(link)
    fd = open(img_path,'wb')
    fd.write(res.content)
    fd.close()
```

```python
download('https://cdn.pixabay.com/photo/2022/03/06/05/30/clouds-7050884__480.jpg', 'img.png')
```

## download all images

```python
paths   = []
img_ids = []

ref = 100000
i = 1

for link in tqdm(df['img_link']):
    img_id = 'A' + str(i + ref)[1:]
    i += 1
    img_path = 'Imgs/' + img_id + '.jpg'
    download(link, img_path)
    
    paths.append(img_path)
    img_ids.append(img_id)
```

```python
df['imd_id']   = img_ids
df['img_path'] = paths

df.head()
```
